```python
def etlSIH(grupo: str, uf: str, ano: int, mes: int or list):
  '''
  Função para execurtar ETL no SIH do DataSUS.

  Variáveis:
  Grupo(String): 'RD': 'AIH Reduzida', 'RJ': 'AIH Rejeitada', 'ER': 'AIH Rejeitada com erro',
 'SP': 'Serviços Profissionais', 'CH': 'Cadastro Hospitalar', 'CM': 'Não informado pela API' .
  UF(String): Sigla dos estados do Brasil
  Ano(Int): 2010 - 2024
  Mês(Int or List): 1-12
  '''
  sih = SIH().load() #api

  #método 1 - escolher grupo, uf, ano e meses
  files = sih.get_files(grupo, uf=uf, year=ano, month=mes)
  data = sih.download(files)

  #método 2 - juntar os datasets baixados no método anterior
  if isinstance(data, list):
    data_list = []
    for i in range(len(data)):
      df = data[i].to_dataframe()
      data_list.append(df)
    df_final = pd.concat(data_list, ignore_index=True)
  else:
    df_final = data.to_dataframe()
  return df_final
```

```python
etlSIH( 'RJ', ["MG", "PE"], 2019, [1])
```
